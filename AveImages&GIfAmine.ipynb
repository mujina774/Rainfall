```python
import glob
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

def ga(k, p, e):
    f = sorted(glob.glob(p + '*' + e))
    images = list(map(lambda file : Image.open(file), f))
    images[0].save(p + pf + str(k).zfill(2) + '.gif', save_all = True, append_images = images[1:], duration = 1200, loop = 0)

def av(k, p, e):
    l = []
    for o in range(1, 13):
        l.append('./im/20' + str(k).zfill(2) + '/' + pf + str(k).zfill(2) + str(o).zfill(2)  + e)
    m = []
    for j in l:
        m.append(np.array(Image.open(j)))  
    m = np.array(m, dtype = 'f8')
    s = m[0, :, :].astype(np.float64)
    for i in range(1, len(m)):
        s += m[i, :, :]
    s = s / 12.
    plt.imshow(s, cmap = "gray")
    plt.imsave(p + pf + str(k).zfill(2) + '.png', s, cmap="gray")
```

```python
y = list(range(1, 24))
ext = '.png'
pf = 'rf20' 
for i in range(0, len(y)):
    path = './im/20' + str(y[i]).zfill(2) + '/'
#    ga(y[i], path, ext)
    av(y[i], path, ext)
```
